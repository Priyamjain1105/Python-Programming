## Class & Object in Python
- Class is a blueprint for creating objects
- Objects are instance of class

```python
class Student:
    name = "karan kumar"
    
s1 = Student()
print(s1.name)

s2 = Student()
print(s2.name)
```

## Constructor `__init__` funtion
- All classes have a function called `__init__()`, which is always executed when the object is being initiated
- Even if we dont explicitely create the init function python it self create the init funtion and executes it
- `init`: initialisation short form

#### Self Parameter
- Constructor always take one parameter which is self parameter
- The self parameter is a reference to teh current instancd of the class, and is used to access variable that belong to the class

##### Concept
- Here fullname is the parameter, and self.name is the variable that is created in the object
- `Good Practise:` Generally programmers give the object variable name same as the parameter

```python
class Student:
    def __init__(self, fullname):
        self.name = fullname
        print(f"Adding new student {self.name} in the database")
        
s1 = Student("karan")
print(s1.name)

s2 = Student("Arjun")
print(s2.name)
```

#### Default constructor and Parameterised Constructor

```python
class Student:
    
    # default constructor
    def __init__(self):
        pass
    
    # Parameterised constructor
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks
        print(f"New student added named {self.name} and marks {self.marks}")
        
s1 = Student("rakesh", 24)
```

### Class Attribute VS Instance Attribute
- `Class attribute:`Those attribute which are same for all objects are class attributes
- `Object Attribute:`Those attributes which change for every object are object attribute

```python
class Student:
    college_name = "VIT"
    def __init__(self, name, marks):
        
        # Class Attribute
       
        
        # Object Attribute
        self.name = name
        self.marks = marks
        print("Adding new student in Database")
    
    
s1 = Student("Priyam", 30)

# Accessing class attribute through obejct
print(s1.college_name)

# Accessing class attribute through class name
print(Student.college_name)
```

**NOTE:** if class attribute and object attribute has same variable name then object attribute has more precedence then class attribute
