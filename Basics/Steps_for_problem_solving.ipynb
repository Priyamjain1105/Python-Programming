## Process to Solve the Problem
1. Analyse your problem
   - Input
   - Ouput 
   - Constraints
2. Break down our problem into smaller subparts
3. Remember/enlist the concept
4. Take 2-3 examples (clear out your underatanding)
5. write a pseudo code on the paper
6. Dry run it once
7. Write down your solution/code
8. Look out for edge cases/bounday condition
