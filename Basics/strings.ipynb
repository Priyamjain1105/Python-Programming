### Game of ASCII Values

```python
print(ord('A')) # returns the ASCII Value of the character
print(chr(65))  # returns the char of the ASCII value
```

### String Slicing

```python
A = "hello_world"
print(A[4:7])
print(A[3:])
print(A[:8])
print(A[4:1:-1])
print(A[9:11])    #out of range will not be printed
print(A[::])
print([A[::-1]])
```

### Operations on Strings


#### Concatenation

#### Matching String

```python
print('r' in 'romeo')
print('J' in 'juliet')
print('paani' in 'jaapaani')
```

#### Comparing String


```python
# Capital letters are smaller then small letters
print('A' == 'A')
print('A'== 'a')
print('abc' == 'abc')
print('ABC'<'XYZ')
```
