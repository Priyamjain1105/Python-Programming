# Create LinkedList

- Here in the first.next we are storing the id of the second
  

```python
class Node:
    def __init__(self, value):
        self.data = value 
        self.next = None
        
first  = Node(1)
second = Node(2)
third = Node(3)

first.next = second
second.next = third

print(id(first), id(second),id(third))
```

Using the delete operation

```python
del(first)
print(first)
```

Accessing the next data


```python
head = first
print(head.data)
print(head.next.data)
print(head.next.next.data)
```

### Printing the linkedlist

```python
temp = head
while(temp !=  None):
    print(temp.data) 
    temp = temp.next
```

### Taking input of the LinkedList


```python
class Node:
    def __inint__(self,value):
        self.data = value
        self.next = None
        
def printLL(head):
    temp = head
    while(temp!=None):
        print(temp.value)
        temp = temp.next
    return

def take_input():
    value = int(input("Enter the value of Node:-"))
    head = None
    
    while(value != -1):
        newNode = Node(value)
        if(head == None):
            head = newNode
        else:
            head.next = newNode
            
        value = int(input("Enter the value of the node:-"))
    return head

newhead = take_input()
printLL(newhead)
```
