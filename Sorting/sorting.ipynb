# Sorting Algorithms


## 1. Selection Sort

```python
def selection_sort(l):
    for i in range(len(l)):
        for j in range(i, len(l)-1):
            if(l[j]<=l[i]):
                l[i],l[j] = l[j], l[i]
    
    for i in l:
        print(i)            

selection_sort([2,5,4,3,8,9,1,6])
```
