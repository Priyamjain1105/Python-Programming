```python
class StackUsingList:
    def __init__(self):
        self.stack=[]
        
    def push(self,data):
        self.stack.append(data)
        print(f"Pushed {data} into Stack")
        
    def size(self):
        return len(self.stack)
    
    def is_empty(self):
        """
        if(len(self.stack)==0):
            return True
        else:
            return False
        """
        return len(self.stack) == 0
    
    def top(self):
        '''
        return self.stack[len(self.stack)-1]
        '''
        if self.is_empty():
            print("Stack is empty, No top element")
            return None
        return self.stack[-1]
    
```
