# Introduction to Function

```python
# Syntax
def function_name(parameters):
    """doc string: to explain the functoion"""
    # function body
    return expression
```

```python
def even_or_odd(num):
    """Function helps to find weather num is even or odd"""
    if num%2==0:
        print("Num is even")
    else:
        print("Num is odd")

## call the function
even_or_odd(23)
```

```python
## function with multiple paramaters
def add(a,b):
    """This function will used to add two variable"""
    c = a+b 
    return c

add(2,4)
```

## Variable length Argument
here we will talk about postional vs variable argument

### Positional Args: 
- we put `*args` to get any number of positional arguments from the function

```python
def print_numbers(*args):
    print(args)
    for i in args:
        print(i)

print_numbers(1,2,3, "ramu")
```

### Keywords Arguments
- In the keywords argument all the values are in teh form of key-value pair

```python
def print_details(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}:{value}")
        
print_details(name="krish", age="32",contry="India")
```

### Combining both of them

```python
def print_details(*args, **kwargs):
    for i in args:
        print("Positional Args:", i)
    
    for key,value in kwargs.items():
        print("Keywarods args:", key,":", value)
    
print_details(1,2,4,5,name="priyam")
```

### Coding Examples


Temperature CONVERSION

```python
def convert(value, unit):
    """This function will convert the temp btw celcius and feharanite"""
    if unit.lower()=='f':
        return value+45
    elif unit.lower()=='c':
        return value*5
    
    
print(convert(46, 'f'))
convert(46,'c')
```

- Password Strength Checker

```python
def is_strong_password(password):
    """This function helps to find the pasword stength"""
    if len(password)<8:
        return False
    if not any(char.isdigit() for char in password):
        return False
    if not any(char.islower() for char in password):
        return False
    if not any(char.isupper() for char in password):
        return False
    if not any(char in '!@#$%^&*()_+' for char in password):
        return False
    
    return True

is_strong_password("ewne")
```
