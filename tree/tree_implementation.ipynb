## Tree Implementation
Trees is a hiarchial data structure consisting of nodes which are connected by edges

```python
class TreeNode:
    def __init__(self,data):
        self.data = data
        self.children = [],

root = TreeNode(1)

child1 = TreeNode(2)
child2 = TreeNode(3)
child3 = TreeNode(4)

root.children.append(child1)
root.children.append(child2)
root.children.append(child3)
```

```python
class TreeNode:
    
    def __init__(self, data):
        self.data = data
        self.children = []
        
    
root = TreeNode(0)
n1 = TreeNode(1)
n2 = TreeNode(2)
n3 = TreeNode(3)

root.children.append(n1)
root.children.append(n2)
n2.children.append(n3)


def print_tree(root):
    print(root.data)
    for eachchild in root.children:
        print_tree(eachchild)
        
print_tree(root)
```

```python
def print_tree(root):
    print(root.data)
    for eachchild in root.children:
        print_tree(eachchild.data)
        
print_tree(root)
```
